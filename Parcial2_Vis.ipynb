```python
import pandas as pd
import numpy as np
```

# Cargar Data

```python
d22 = pd.read_excel('DATOS ABIERTOS JULIO 2022.xlsx', index_col=0)  
```

```python
d21 = pd.read_excel('DATOS_ABIERTOS_2021[2].xlsx', index_col=0)  
```

# Limpiar Data 2021

```python
d21 = d21[d21['FECHA_FIN']!='-']
d21 = d21[d21["COSTO_BIEN"]!='-']
```

# Limpiar Data 2022

```python
d22.info()
```

```python
d22.isna().sum()
```

```python
d22 = d22.dropna()
```

```python
d22 = d22[d22['FECHA_FIN']!='-']
d22 = d22[d22["COSTO_BIEN"]!='-']
```

```python
d22 = d22.drop('FECHA DE CIERRE', axis=1)
```

```python
d22['FECHA_FIN'] = pd.to_datetime(d22['FECHA_FIN'])
```

```python
d22['Month'] = (d22['FECHA_INGRESO']).dt.month
d22.head()
```

```python
d22['Month'].unique()
```

```python
d22.info()
```

## Hay filtrar los datos de 2021, para que coincidan los meses.

```python
d21 = d21[d21['Month'] <= 5]
```

```python
d21['Month'].unique()
```

```python
d22['Month'].unique()
```

# Crear Metricas

```python
d21["TIEMPO_RESOLUCION_DIAS"] =(d21['FECHA_FIN'] - d21['FECHA_INGRESO']) / np.timedelta64(1, 'D')
```

```python
d21.head()
```

```python
d22["TIEMPO_RESOLUCION_DIAS"]=(d22['FECHA_FIN'] - d22['FECHA_INGRESO']) / np.timedelta64(1, 'D')
```

```python
d22.head()
```

# Analisis

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
sectores21 = d21.groupby('SECTOR')
sectores21.count()
```

```python
sectores22 = d22.groupby('SECTOR')
sectores22.count()
```

```python
tipos21 = d21.groupby('TIPO_RECLAMACION')
tipos21.count()
```

```python
tipos22 = d22.groupby('TIPO_RECLAMACION')
tipos22.count()
```

```python
toptipos21 = tipos21['FECHA_INGRESO'].count().nlargest(n=5).to_frame()
toptipos21 = toptipos21.sort_values(by = 'FECHA_INGRESO',ascending = True)
toptipos21 = toptipos21.reset_index(level=0)
toptipos21 = toptipos21.rename(columns={"FECHA_INGRESO": "2021"})
```

```python
toptipos22 = tipos22['FECHA_INGRESO'].count().nlargest(n=5).to_frame()
toptipos22 = toptipos22.sort_values(by = 'FECHA_INGRESO',ascending = True)
toptipos22 = toptipos22.reset_index(level=0)
toptipos22 = toptipos22.rename(columns={"FECHA_INGRESO": "2022"})
```

```python
toptipos = toptipos22
toptipos['2021'] = toptipos21['2021']
toptipos=toptipos.set_index('TIPO_RECLAMACION')
toptipos
```

```python
ax = toptipos.plot.barh(color={"2021": "blue", "2022": "orange"}, title = 'Top 5 Tipo de Reclamos 2021 vs 2022')
ax.set_xlabel('Numero de Quejas')
ax.set_ylabel('Tipo de Reclamo')
plt.show();
```

```python
sectores21 = d21.groupby(['SECTOR'])
print(sectores21['SECTOR'].count().nlargest(n=7))
```

```python
sec = ['Inmobiliario', 'Mueblero', 'Compañia de autofinanciamiento', 'Automotriz', 'Tienda deparmental', 'Electricidad', 'Turístico']
plt.barh(sec, sectores21['SECTOR'].count().nlargest(n=7).sort_values())
plt.ylabel('SECTORES')
plt.xlabel('NUM DE QUEJAS')
plt.show()
```

```python
sectores22 = d22.groupby('SECTOR')
print(sectores22['SECTOR'].count().nlargest(n=7))
```

```python
sec2 = ['Automotriz', 'Taller mecánico', 'Ventas por internet', 'Tienda deparmental', 'Electricidad', 'Tiendas de autoservicio', 'Turístico']
plt.barh(sec2, sectores22['SECTOR'].count().nlargest(n=7).sort_values())
plt.ylabel('SECTORES')
plt.xlabel('NUM DE QUEJAS')
plt.show()
```

```python
d21["TIEMPO_RESOLUCION_DIAS"].hist(width=40,edgecolor='black');
d22["TIEMPO_RESOLUCION_DIAS"].hist(width=15,edgecolor='black');
plt.ylabel('Frecuencia');
plt.xlabel('Tiempo de Resolucion en dias');
plt.title('Histograma de Tiempo de Resolucion 2021 vs 2022');
plt.grid(visible=None)
plt.legend(['2021','2022']);
```
